```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv('./iris.csv')
```

```python
iris_data = iris.iloc[:, 0:4].values
iris_target = iris.iloc[:, 4].values
```

```python
# As classes devem ser representadas por 3 numeros int (0 ou 1) para as colunas setosa, virginica e versicolor
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
labelencoder = LabelEncoder()
indexes = labelencoder.fit_transform(iris_target)
categories = to_categorical(indexes)
```

```python
# 0 setosa
# 1 virginica
# 2 versicolor
print(indexes)
# 1 0 0 setosa
# 0 1 0 virginica
# 0 0 1 versicolor
print(categories)
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, target_train, target_test = train_test_split(iris_data, categories, test_size=0.30)
```

## Parâmetros importantes (https://keras.io/api/)
- input_dim : quantidade de atributos de entrada. Aparece apenas na primeira camada 
- units : neuronios de cada entrada
- activation : função de ativação, vários tipos para camadas intermediarias. softmax na ultima camada para classificação de 3 ou mais classes. Em classificação binária, usar um neuronio (units=1) na última camada com função sigmoid
- optimizer : testar mais de um, pode iniciar com adam
- loss : categorical_crossentropy para mais 3 ou mais classes, binary_entropy para classificação binária
- metrics : categorical_accuracy pode ser usada para 3 ou mais classes e binary_accuracy para classificação binária
- mas há outras métricas

```python
from tensorflow.keras.models import Sequential # Estrutura de rede
from tensorflow.keras.layers import Dense      # camadas densas

classifier = Sequential()
# Tamanho da primeira camada (<entradas> + <saidas>) /2 = (4 + 3)/2 = 3.5 --> 4 (arredonda pra cima) 
classifier.add(Dense(units=4, activation='relu', input_dim=4)) #input_dim = número de atributos

# units = 3 para 3 classes
classifier.add(Dense(units=3, activation='softmax')) # softmax para 3 ou mais classes

classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])

classifier.fit(data_train, target_train, batch_size=10, epochs=500)
```

```python
predictions = classifier.predict(data_test)

print(predictions)
```

```python
# Transforma em true scores maiores que 0.5 e em false scores menores ou iguais a 0.5
predictions = (predictions > 0.5)

import numpy as np
# retorna o indice (0,1 ou 2) com maior score para cada instancia das das classes alvo (gabarito) e das predições
target_test_index = [np.argmax(t) for t in target_test] 
predictions_index = [np.argmax(t) for t in predictions]

print(target_test_index)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(target_test_index, predictions_index))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(target_test_index, predictions_index))
```

```python
from sklearn.metrics import classification_report
print(classification_report(target_test_index, predictions_index))
```

# Validação Cruzada (Cross Validation)

```python
from tensorflow.keras.models import Sequential # Estrutura de rede
from tensorflow.keras.layers import Dense      # camadas densas
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import numpy as np


def get_classifier():
    classifier = Sequential()
    # Tamanho da primeira camada (<entradas> + <saidas>) /2 = (4 + 3)/2 = 3.5 --> 4 (arredonda pra cima) 
    classifier.add(Dense(units=4, activation='relu', input_dim=4)) #input_dim = número de atributos

    # units = 3 para 3 classes
    classifier.add(Dense(units=3, activation='softmax')) # softmax para 3 ou mais classes

    classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
    
    return classifier


kfold = KFold(n_splits=10, shuffle=True)

accuracies = []

# utiliza o dataset inteiro iris_data (atributos) e categories (classes)
for train_index, test_index in kfold.split(iris_data):
    X_train, X_test = iris_data[train_index], iris_data[test_index]
    y_train, y_test = categories[train_index], categories[test_index]

    model = get_classifier()
    model.fit(x=X_train, y=y_train, batch_size=10, epochs=500, verbose=0)

    y_pred = model.predict(X_test)
    
    predictions = (y_pred > 0.5)

    # retorna o indice (0,1 ou 2) com maior score para cada instancia das das classes alvo (gabarito) e das predições
    target_test_index = [np.argmax(t) for t in y_test] 
    predictions_index = [np.argmax(t) for t in predictions]

    accuracy = accuracy_score(target_test_index, predictions_index)
    accuracies.append(accuracy)
    print(f'Accuracy: {accuracy}')

print('Accuracia Média', np.mean(accuracies))

```
