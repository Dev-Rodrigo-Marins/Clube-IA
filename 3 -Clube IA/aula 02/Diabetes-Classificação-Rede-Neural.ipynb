```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
diabetes = pd.read_csv('./diabetes.csv')
diabetes.shape
```

```python
classes = ['tested_negative', 'tested_positive']
```

```python
diabetes.head()
```

```python
diabetes_data = diabetes.iloc[:, 0:8].values
diabetes_target = diabetes.iloc[:, 8].values
```

```python
diabetes_data
```

```python
from sklearn.preprocessing import label_binarize

diabetes_target_num = label_binarize(diabetes_target, classes=classes)

print(diabetes_target_num)
```

```python
diabetes_target = np.array([ x[0] for x in diabetes_target_num])
```

```python
diabetes_target
```

```python
diabetes_data
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, target_train, target_test = train_test_split(diabetes_data, diabetes_target, test_size=0.20)
```

```python
from tensorflow.keras.models import Sequential # Estrutura de rede
from tensorflow.keras.layers import Dense      # camadas densas
from tensorflow.keras.metrics import Recall

classifier = Sequential()
# Tamanho da primeira camada (<entradas> + <saidas>) /2 = (8 + 1)/2 = 4.5 --> 5 (arredonda para cima)
classifier.add(Dense(units=5, activation='relu', input_dim=8)) #input_dim = número de atributos
classifier.add(Dense(units=1, activation='sigmoid')) # 1 neuronio com sigmoid para 2 classes (binário)

classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])
# se quiser priorizar o recall na classe positica

#classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=[Recall()])

classifier.fit(data_train, target_train, batch_size=30, epochs=2000)
```

```python
predictions = classifier.predict(data_test)

print(predictions)
```

```python
# Transforma em true scores maiores que 0.5 e em false scores menores ou iguais a 0.5
predictions = list(map(int,predictions > 0.5))
```

```python
predictions
```

```python
target_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(target_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_true=target_test, y_pred=predictions))
```

```python
from sklearn.metrics import classification_report
print(classification_report(target_test, predictions))
```

# Validação Cruzada

```python
from tensorflow.keras.models import Sequential # Estrutura de rede
from tensorflow.keras.layers import Dense      # camadas densas
from sklearn.model_selection import KFold
from sklearn.metrics import recall_score
import numpy as np

def get_classifier():
    classifier = Sequential()
    # Tamanho da primeira camada (<entradas> + <saidas>) /2 = (8 + 1)/2 = 4.5 --> 5 (arredonda para cima)
    classifier.add(Dense(units=5, activation='relu', input_dim=8)) #input_dim = número de atributos
    classifier.add(Dense(units=1, activation='sigmoid')) # 1 neuronio com sigmoid para 2 classes (binário)

    #classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['binary_accuracy'])
    classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=[Recall()])
    
    return classifier


kfold = KFold(n_splits=5, shuffle=True)

recalls = []

# utiliza o dataset inteiro diabetes_data (atributos) e diabetes_target (classes)
for train_index, test_index in kfold.split(diabetes_data):
    X_train, X_test = diabetes_data[train_index], diabetes_data[test_index]
    y_train, y_test = diabetes_target[train_index], diabetes_target[test_index]

    model = get_classifier()
    model.fit(x=X_train, y=y_train, batch_size=30, epochs=2000, verbose=0)

    y_pred = model.predict(X_test)
    
    predictions = (y_pred > 0.5)

    recall = recall_score(y_test, predictions)
    recalls.append(recall)
    print(f'Recall: {recall}')

print('Recall Médio', np.mean(recalls))

```
