```python
import serial
import csv

# Cria uma conexão serial (Substitua pela sua porta serial)
arduino = serial.Serial('/dev/cu.usbmodem14701', 9600)  
print("Conexão serial aberta.")

arduino.flushInput() # Limpa o buffer de entrada

# Cria uma matriz para armazenar os dados
ldr_dados = []

# Lê 150 linhas da serial
for i in range(150):
    
    linha = arduino.readline().decode('utf-8').strip()  # Lê uma linha
                                                       # remove espaços em branco 
                                                      # e quebras de linha
    print(i, linha)         # Mostra a linha lida
    valores = linha.split()  # Divide a linha em valores separados por espaços
    ldr_dados.append(valores)  # Adiciona os valores à matriz

# Fecha a conexão serial
arduino.close()
print("Conexão serial fechada.")

# Salva os dados em um arquivo CSV
with open('dados_ldr.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(ldr_dados)
print("Dados salvos em ldr_dados.csv.")
```
